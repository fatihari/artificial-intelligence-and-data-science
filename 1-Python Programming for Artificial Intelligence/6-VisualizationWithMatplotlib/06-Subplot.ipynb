```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget

df = pd.read_csv("dataset\iris.csv") # df=dataFrame
df1 = df.drop(["Id"],axis=1) # id bilgisi kullancağım bir veri olmadığı için drop ederiz.
df1
```

```python
df1.plot(grid=True,alpha= 0.9,subplots = True)
```

```python
setosa = df[df.Species == "Iris-setosa"]
versicolor = df[df.Species == "Iris-versicolor"]
virginica = df[df.Species == "Iris-virginica"]
```

```python
plt.title("PetalLengthCm")
plt.subplot(3,1,1) # (3 tane subplot, 1. column, 1. satır)
plt.plot(setosa.Id,setosa.PetalLengthCm,color="red",label= "setosa")
plt.ylabel("setosa")

plt.subplot(3,1,2) # (3 tane subplot, 1. column, 2. satır)
plt.plot(versicolor.Id,versicolor.PetalLengthCm,color="green",label= "versicolor")
plt.ylabel("versicolor")

plt.subplot(3,1,3) # (3 tane subplot, 1. column, 3. satır)
plt.plot(versicolor.Id,versicolor.PetalLengthCm,color="blue",label= "versicolor")
plt.ylabel("virginica")
plt.show()
```
