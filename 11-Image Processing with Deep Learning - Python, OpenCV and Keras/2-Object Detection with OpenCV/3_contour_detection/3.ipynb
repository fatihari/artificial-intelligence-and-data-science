```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## 1-Read Image

```python
img = cv2.imread("img/contour.jpg",0) # 0=siyah beyaz
plt.figure(figsize=(9,9)), plt.imshow(img, cmap = "gray"), plt.axis("off")
plt.show()
```

## 2-Find Contour

```python
contours, hierarch = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
external_contour = np.zeros(img.shape)
internal_contour = np.zeros(img.shape)
```

```python
for i in range(len(contours)):
    
    # external
    if hierarch[0][i][3] == -1:
        cv2.drawContours(external_contour,contours, i, 255, -1) # -1 konturlarla sınırlı olan alanı doldur demek => normalde kalınlığı belirtir.
    else: # internal
        cv2.drawContours(internal_contour,contours, i, 255, -1)
```

```python
plt.figure(figsize=(9,9)), plt.imshow(img, cmap = "gray"), plt.axis("off")
plt.figure(figsize=(9,9)), plt.imshow(external_contour, cmap = "gray"), plt.axis("off")
plt.figure(figsize=(9,9)), plt.imshow(internal_contour, cmap = "gray"), plt.axis("off")
plt.show()
```
