```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## 1-Read Image

```python
img = cv2.imread("img/sudoku.jpg", 0)
img = np.float32(img)
print(img.shape)
plt.figure(figsize=(16,9)), plt.title("Sudoku"), plt.imshow(img, cmap = "gray"), plt.axis("off")
plt.show()
```

## 2-Harric Corner Detection

```python
# blockSize = komşuluk boyutu
dst = cv2.cornerHarris(src = img, blockSize = 2, ksize = 3, k = 0.04)
plt.figure(figsize=(16,9)), plt.imshow(dst, cmap = "gray"), plt.axis("off")
plt.show()
```

## 3-Dilation of Detected Points

```python
dst = cv2.dilate(dst, None) # genişletme methodu (Dilation)
img[dst > 0.2 * dst.max()] = 1
plt.figure(figsize=(16,9)), plt.imshow(dst, cmap = "gray"), plt.axis("off")
plt.show()
```

```python
img = cv2.imread("img/sudoku.jpg", 0)
img = np.float32(img)
print(img.shape)
plt.figure(figsize=(16,9)), plt.title("Sudoku"), plt.imshow(img, cmap = "gray"), plt.axis("off")
plt.show()
```

```python
feature_params = dict(  maxCorners = 120, 
                        qualityLevel = 0.01, 
                        minDistance = 10)

corners = cv2.goodFeaturesToTrack(img, **feature_params)
corners = np.int64(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(img, (x,y),3,(125,125,125),cv2.FILLED)
    
plt.figure(figsize=(16,9)), plt.title("Sudoku"), plt.imshow(img), plt.axis("off")
plt.show()
```
