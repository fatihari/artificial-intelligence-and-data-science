```python
import keyboard # pip install keyboard
import uuid # for recording from screen
import time
from PIL import Image # Python Image Library
from mss import mss # pip install mss
```

```python
position = {"top": 460, "left": 700, "width": 300, "height": 125}
sct = mss() # yukarıdaki konumu kesecek ve frame haline dönüştürecek.
```

- **key** = keyboard command => Up and down arrow keys
- **i** = number of presses

```python
i = 0

def record_screen(record_id, key):
    global i # hem içerde hem dışarda kullanabilmek için i global oldu.
    
    i += 1
    print("{}: {}".format(key, i)) 
    img = sct.grab(position) # catch pixels in spesific position
    im = Image.frombytes("RGB", img.size, img.rgb) # read as RGB format
    im.save("./img/{}_{}_{}.png".format(key, record_id, i))
```

```python
is_exit = False

def exit():
    global is_exit
    is_exit = True  # esc tuşuna basılınca veri toplama işlemi bitecek.

keyboard.add_hotkey("esc", exit) 
```

```python
record_id = uuid.uuid4()

while True:
    
    if is_exit: break # ilk çıkıp çıkılmadığı kontrol edilir.

    try:
        if keyboard.is_pressed(keyboard.KEY_UP):
            record_screen(record_id, "up") # key = up
            time.sleep(0.1) # delay
        elif keyboard.is_pressed(keyboard.KEY_DOWN):
            record_screen(record_id, "down") # key = down
            time.sleep(0.1) # delay
        elif keyboard.is_pressed("right"): # sağa basınca zıplamayacak veya eğilmeyecek, sağa doğru ilerleyecek.
            record_screen(record_id, "right") # key = right
            time.sleep(0.1) # delay
    except RuntimeError: continue # try'da herhangi bir hatada, kodu devam ettir, ancak runtime error da ver.
```
