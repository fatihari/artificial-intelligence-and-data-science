### NOTE:
- **step**: diktörtgenin resim üzerinde kaç piksel atlayarak dolaşacağını belirtir. 
- 1 seçilirse, 1 pixel kaya kaya dolaşacaktır.
- **ws**: dikdörtgenin boyutu

```python
def sliding_window(image, step, ws):
    
    for y in range(0, image.shape[0]-ws[1], step): # step aralıklarla artar (0, 200, 400)
        for x in range(0, image.shape[1]-ws[0],step):
            
            yield(x,y, image[y:y+ws[1], x:x+ws[0]])
```
