```python
import cv2
import matplotlib.pyplot as plt
```

### NOTE
- Resimleri **list**'in içerisinde depolamak yerine anlık üretmek için **yield** kullanılır.
- Bu şekilde memory'den kazanç sağlanır.

### NOTE

- minSize = (224, 224),
  -  minSize[0] =>  height
  - minSize[1] =>  width


- image.shape[0] => width
- image.shape[1] => height

```python
def image_pyramid(image, scale = 1.5, minSize = (224, 224)):
    
    yield image 
    
    while True:
        
        width = int(image.shape[1]/scale) # 886/1.5=590.66 => int(590.66) = 590
        height = width
        image = cv2.resize(image, dsize=(width, height))
        
        if image.shape[0] < minSize[1] or image.shape[1] < minSize[0]:
            break
        
        yield image # return image with generators
```
