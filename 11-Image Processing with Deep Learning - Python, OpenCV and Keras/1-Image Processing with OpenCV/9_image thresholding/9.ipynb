```python
import cv2
import matplotlib.pyplot as plt
```

## 1-Import Image

```python
img = cv2.imread("img1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert gray
```

```python
%matplotlib qt5
plt.figure()
plt.imshow(img, cmap = "gray")
plt.axis("off")
plt.show()
```

```python
threshold = 60
```

- threshold methodu 2 parametre döner, birisi bizim için önemli değil, o yüzden **_** koyarız.
- threshold örneğin 60 olsun. Bu değerin üzerindeki değerleri **THRESH_BINARY** parametresi ile beyaz yapar yani 60'ın üzerindeki değerleri göstermez.
- Eğer 60'ın altındaki değerleri siyah yapsın siyahları da beyaz yapsın istiyorsak: **THRESH_BINARY_INV** kullanabiliriz.

```python
_, thresh_img = cv2.threshold(img, thresh = threshold, maxval = 255, type = cv2.THRESH_BINARY)
```

```python
%matplotlib inline
plt.figure()
plt.imshow(thresh_img, cmap = "gray")
plt.axis("off")
plt.show()
```

- Bazı bölümlerde adaptive threshold kullanılması gerekir.
- Çünkü örneğin bir tepeye düşen ışık miktarı kimi tarafta az kimi tarafta fazla olduğu için fazla ışıklı olan beyaz kısmı kaybetmiş oluyoruz. 
- Bunu engellemek için adaptive threshold kullanılır.

- The **blockSize** determines **the size of the neighbourhood area.**
- **C** is a constant that is subtracted from the **mean** or **weighted sum of the neighbourhood pixels**.

```python
thresh_img2 = cv2.adaptiveThreshold(src = img, maxValue= 255, adaptiveMethod =  cv2.ADAPTIVE_THRESH_MEAN_C, thresholdType = cv2.THRESH_BINARY, blockSize = 11, C = 8)
```

```python
plt.figure()
plt.imshow(thresh_img2, cmap = "gray")
plt.axis("off")
plt.show()
```

```python
thresh_img3 = cv2.adaptiveThreshold(src = img, maxValue= 255, adaptiveMethod =  cv2.ADAPTIVE_THRESH_GAUSSIAN_C, thresholdType = cv2.THRESH_BINARY, blockSize = 11, C = 8)
```

```python
plt.figure()
plt.imshow(thresh_img3, cmap = "gray")
plt.axis("off")
plt.show()
```
