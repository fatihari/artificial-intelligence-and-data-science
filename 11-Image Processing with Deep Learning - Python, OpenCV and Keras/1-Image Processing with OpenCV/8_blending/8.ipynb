```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("img1.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) # renk skalasını bgr'den rgb'ye dönüştürülür.
img2 = cv2.imread("img2.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.figure()
plt.imshow(img1)

plt.figure()
plt.imshow(img2)
```

```python
print(img1.shape)
print(img2.shape)
```

- Resimleri karıştırmadan önce boyutlarının aynı olması gerekir. 
- İki resmin boyutunu da 600x600 yapıyoruz.

```python
img1 = cv2.resize(img1, (600,600))
print(img1.shape)

img2 = cv2.resize(img2, (600,600))
print(img2.shape)
```

```python
plt.figure()
plt.imshow(img1)

plt.figure()
plt.imshow(img2)
```

```python
# karıştırılmış resim(addWeighted) = alpha*img1 + beta*img2
blended = cv2.addWeighted(src1 = img1, alpha =0.5, src2= img2, beta = 0.5, gamma = 0)
```

```python
plt.figure()
plt.imshow(blended)
```
